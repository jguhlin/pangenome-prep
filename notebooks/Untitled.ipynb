```python
import polars as pl
from sklearn.decomposition import PCA
import plotly.express as px
```

```python
df = pl.read_csv("../contamination_matrix.tsv", has_header=False, separator="\t")
df.head()
```

```python
pca = PCA(n_components=3)
pca.fit(df[:, 1:-1].to_numpy())
x = pca.transform(df[:, 1:-1].to_numpy())
```

```python
px.scatter(x[:, 0], x[:, 1])
```
